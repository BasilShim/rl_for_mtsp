```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

path = '/home/basilshim/ray_results/ppo_n10/PPOTrainer_TransportScape_2022-05-29_06-45-18yprxlewa/progress.csv'
df = pd.read_csv(path)
re = df.values[:,2]
t = df.values[:,6]
figure(figsize=(9, 6))
plt.plot(t, re, color="black")
plt.xlabel("t")
plt.ylabel("Награда")
#plt.autoscale(True)
plt.xticks(np.arange(0, t[-1], 100000))
plt.legend()
plt.show()
```
