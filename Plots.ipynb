```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib.ticker import FuncFormatter
import numpy as np

def mills(x, pos):
    return "%.1f M" % (x*1e-6)

formatter = FuncFormatter(mills)

path = '/home/basilshim/ray_results/PPOTrainer_TransportScape_2022-06-08_12-58-254_zntdfn/progress.csv'
df = pd.read_csv(path)
re = df.values[:,2]
t = df.values[:,6]
fig, ax = plt.subplots()
fig.set_figheight(6)
fig.set_figwidth(9)
ax.plot(t, re, color="black")
ax.set_xlabel("t")
ax.set_ylabel("Награда")
ax.xaxis.set_major_formatter(formatter)
#plt.autoscale(True)
ax.legend()
plt.show()
```
